Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year
of release) and make data frame. 

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=2e9dfa9b-3e4d-4d39-acd2-8af11f252a59&pf_rd_r=5GSJ3ES3VHSWGNDF24QZ&pf_rd_s=right-5&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_india_tr_rhs_1')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
name=soup.find_all("td",class_="titleColumn")
name
```

```python
movie_name=[]
for i in name:
  movie_name.append(i.text.replace('\n',''))
movie_name[:100]
```

```python
movie_name=[]
for i in name:
  for j in i.find_all('a'):
    movie_name.append(j.text.replace('\n',''))
movie_name
```

```python
year=soup.find_all("span",class_="secondaryInfo")
movie_year=[]
for i in year:
  movie_year.append(i.text.replace('\n',''))
movie_year[:100]
```

```python
rating=soup.find_all("td",class_="ratingColumn imdbRating")
movie_rating=[]
for i in rating:
  movie_rating.append(i.text.replace('\n',''))
movie_rating[:100]
```

```python
import pandas as pd
Movies=pd.DataFrame({})
Movies['movie_name']=movie_name
Movies['movie_year']=movie_year
Movies['movie_rating']=movie_rating
```

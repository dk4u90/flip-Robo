```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Amazon.in
driver.get('https://github.com/')
```

```python
x=driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details/div/ul[2]/li[3]/a").get_attribute('href')
driver.get(x)
```

```python
Repository_title=[]
Repository_description=[]


# Extracting Repository_title from xpath
try:
    for i in driver.find_elements_by_xpath('//span[@class="text-normal"]'):
        Repository_title.append(i.text)
except NoSuchElementException:
    Repository_title.append("_")


# Extracting Repository_description from xpath
try:
    for i in driver.find_elements_by_xpath('//p[@class="col-9 color-text-secondary my-1 pr-4"]'):
        Repository_description.append(i.text)
except NoSuchElementException:
    Repository_description.append("_")
```

```python
GitHub=pd.DataFrame({})
GitHub['Repository_title']=Repository_title[0:20]
GitHub['Repository_description']=Repository_description[0:20]
GitHub
```

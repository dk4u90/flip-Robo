```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Wikipedia.
driver.get('https://economictimes.indiatimes.com/archive/year-2018,month-4.cms')
```

```python
Date_URL=[]
try:
    #scraping Date_URL
    for i in driver.find_elements_by_xpath("//table[@id='calender']/tbody/tr/td/a"):
        #appending the urls in Date_URL list
        Date_URL.append(i.get_attribute('href'))
except NoSuchElementException :
    Date_URL.append("Not Available")
```

```python
Headline_URL=[]
for j in Date_URL:
    driver.get(j)
    try:
        #scraping Headline_URL
        for i in driver.find_elements_by_xpath("//section[@id='pageContent']/span/table[2]/tbody/tr/td/ul/li/a"):
            #appending the urls in Headline_URL list
            Headline_URL.append(i.get_attribute('href'))
    except NoSuchElementException :
        Headline_URL.append("Not Available")
```

```python
len(Headline_URL)
```

```python
Headline_URL
```

```python
#for loop for scrapping data.
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
for url in Headline_URL:
    driver.get(url)
    time.sleep(2)
    # Extracting Headline from xpath
    try: 
        headline=driver.find_element_by_xpath("//div[@class='article_wrap']/div[1]/div/h1").text
        Headline.append(headline)
    except NoSuchElementException :
        Headline.append("Not Available")
        
    # Extracting Vertical from xpath       
    try: 
        vertical=driver.find_element_by_xpath("//div[@class='clr breadCrumb contentwrapper']/span[3]/a").text
        Vertical.append(vertical)
    except NoSuchElementException :
        Vertical.append("Not Available")
        
    # Extracting Author from xpath
    try: 
        author=driver.find_element_by_xpath("//div[@class='article_wrap']/div[1]/div[2]/div[1]/div/span[1]").text
        Author.append(author)
    except NoSuchElementException :
            Author.append("Not Available")
            
    # Extracting Date from xpath
    try: 
        date=driver.find_element_by_xpath("//div[@class='article_wrap']/div[1]/div[2]/div/div/time").text
        Date.append(date)
    except NoSuchElementException :
        Date.append("Not Available")
            
    # Extracting Description from xpath
    try: 
        description=driver.find_element_by_xpath("//div[@class='article_wrap']/div[1]/div[3]/div/article/div[2]").text
        Description.append(description)
    except NoSuchElementException :
        Description.append("Not Available")
```

```python
#create Dataframe.
The_Economic_Times=pd.DataFrame({})
The_Economic_Times['Date']=Date
The_Economic_Times['Author']=Author
The_Economic_Times['Vertical']=Vertical
The_Economic_Times['Headline']=Headline
The_Economic_Times['Description']=Description
The_Economic_Times
```

```python
# saving the dataframe
The_Economic_Times.to_csv('TET.csv')
```

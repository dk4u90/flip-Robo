Q1: Write a python program to scrape data for “Data Analyst” Job position in 
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.

```python
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
location=[]
for i in location_tags:
    location.append(i.text)
location
```

Scrap Locations

```python
job_discriptions=driver.find_elements_by_xpath("//div[@class='job-description fs12 grey-text']")
discriptions=[]
for i in job_discriptions:
    discriptions.append(i.text)
discriptions
```

Scrapped descriptions Data and stored in list.

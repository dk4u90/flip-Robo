Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day
extended fore

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YK1i9agzZPZ')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
d1 = soup.find_all("p", class_ ="period-name")
Day = []
for i in d1:
 Day.append(i.text)
Day
```

```python
t1 = soup.find_all('p', class_='temp temp-high')
Temp = []
for i in t1:
 Temp.append(i.text)
Temp
```

```python
d2 = soup.find_all("p", class_="short-desc")
Description = []
for i in d2:
 Description.append(i.text)
Description
```

```python
import pandas as pd
Weather = pd.DataFrame()
Weather['Day'] = Day[0:5]
Weather['Temp'] = Temp[0:5]
Weather['Description'] = Description[0:5]
Weather
```

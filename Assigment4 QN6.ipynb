```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Amazon.in
driver.get('https://www.billboard.com/')
```

```python
driver.find_element_by_xpath('//ul[@class="header__subnav__list display--flex flex--left-center flex-md--center-center"]/li[3]').click()
```

```python
Song_name=[]
Artist_name=[]
Last_week_rank=[]
Peak_rank=[]
Weeks_on_board=[]

# Extracting Song_name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="chart-list container "]/ol/li/button/span[2]/span[1]'):
        Song_name.append(i.text)
except NoSuchElementException:
    Song_name.append("_")
    
# Extracting Artist_name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="chart-list container "]/ol/li/button/span[2]/span[2]'):
        Artist_name.append(i.text)
except NoSuchElementException:
    Artist_name.append("_")
    
# Extracting Last_week_rank from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="chart-list container "]/ol/li/button/div/div[2]'):
        Last_week_rank.append(i.text)
except NoSuchElementException:
    Last_week_rank.append("_")
    
    
# Extracting Peak_rank from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="chart-list container "]/ol/li/button/div/div[3]'):
        Peak_rank.append(i.text)
except NoSuchElementException:
    Peak_rank.append("_")
    
    

# Extracting Weeks_on_board from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]'):
        Weeks_on_board.append(i.text)
except NoSuchElementException:
    Weeks_on_board.append("_")
```

```python
Billboard=pd.DataFrame({})
Billboard['Song_name']=Song_name
Billboard['Artist_name']=Artist_name
Billboard['Last_week_rank']=Last_week_rank
Billboard['Peak_rank']=Peak_rank
Billboard['Weeks_on_board']=Weeks_on_board
Billboard
```

```python
import time
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.guru99.com/')
```

```python
driver.find_element_by_xpath('//div[@class="row featured-boxes"]/div/div/ul[1]/li[3]/a').click()
```

```python
driver.find_element_by_xpath('//div[@class="item-page"]/div[2]/table[5]/tbody/tr[34]/td/a').click()
```

```python
Name=[]
Description=[]


# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath('//table[@class="table table-striped"]/tbody/tr/td[1]'):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("_")
    
    
# Extracting Description from xpath
try:
    for i in driver.find_elements_by_xpath('//table[@class="table table-striped"]/tbody/tr/td[2]'):
        Description.append(i.text)
except NoSuchElementException:
    Description.append("_")
    
```

```python
Guru99=pd.DataFrame({})
Guru99['Name']=Name
Guru99['Description']=Description
```

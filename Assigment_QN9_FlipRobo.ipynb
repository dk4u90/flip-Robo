Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,
company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://internshala.com/fresher-jobs')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title=soup.find_all('div',class_='heading_4_5 profile')
title
```

```python

title_name=[]
for i in title:
  title_name.append(i.text.replace('\n',''))
title_name
```

```python
company=soup.find_all('a',class_='link_display_like_text')
company
```

```python
company_name=[]
for i in company:
  company_name.append(i.text.strip())
company_name
```

```python
location=soup.find_all('a',class_='location_link')
location
```

```python
company_location=[]
for i in location:
  company_location.append(i.text.replace('\n',''))
company_location
```

```python
get_ctcadate = soup.find_all('div',class_ = "item_body", id = False)
get_ctcadate
```

```python
ctc=[]
apply_date=[]
temp=[]
for cdate in get_ctcadate:
        temp.append(cdate.text.strip())
temp
```

```python
for i in temp:
  ctc.append(i) if 'LPA' in i else apply_date.append(i)
```

```python
apply_date
```

```python
print(len(company_name),len(ctc),len(title_name),len(apply_date))
```

```python
import pandas as pd
internshala= pd.DataFrame({})
internshala['Company']=company_name
internshala['Title']=title_name
internshala['CTC']=ctc
internshala['Apply Date']=apply_date
internshala
```

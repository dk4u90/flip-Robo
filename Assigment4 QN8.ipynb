```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Amazon.in
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/')
```

```python
Book_name=[]
Author_name=[]
Volumes_sold=[]
Publisher=[]
Genre=[]


# Extracting Book_name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="embed block"]/table/tbody/tr/td[2]'):
        Book_name.append(i.text)
except NoSuchElementException:
    Book_name.append("_")
    
# Extracting Author_name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="embed block"]/table/tbody/tr/td[3]'):
        Author_name.append(i.text)
except NoSuchElementException:
    Author_name.append("_")
    
# Extracting Volumes_sold from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="embed block"]/table/tbody/tr/td[4]'):
        Volumes_sold.append(i.text)
except NoSuchElementException:
    Volumes_sold.append("_")
    
    
# Extracting Publisher from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="embed block"]/table/tbody/tr/td[5]'):
        Publisher.append(i.text)
except NoSuchElementException:
    Publisher.append("_")
    
# Extracting Genre from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="embed block"]/table/tbody/tr/td[6]'):
        Genre.append(i.text)
except NoSuchElementException:
    Genre.append("_")
```

```python
Novels=pd.DataFrame({})
Novels['Book_name']=Book_name
Novels['Author_name']=Author_name
Novels['Volumes_sold']=Volumes_sold
Novels['Publisher']=Publisher
Novels['Genre']=Genre
Novels
```

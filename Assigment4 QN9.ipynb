```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Amazon.in
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
Name=[]
Year=[]
Votes=[]
Genre=[]
Ratings=[]
Run_time=[]


# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a'):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("_")
    
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/span[2]'):
        Year.append(i.text)
except NoSuchElementException:
    Year.append("_")
    
# Extracting Ratings from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]'):
        Ratings.append(i.text)
except NoSuchElementException:
    Ratings.append("_")
    
    
# Extracting Run_time from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p/span[3]'):
        Run_time.append(i.text)
except NoSuchElementException:
    Run_time.append("_")
    
# Extracting Genre from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p/span[5]'):
        Genre.append(i.text)
except NoSuchElementException:
    Genre.append("_")
    

# Extracting Ratings from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p[4]/span[2]'):
        Votes.append(i.text)
except NoSuchElementException:
    Votes.append("_")
```

```python
MBD=pd.DataFrame({})
MBD['Name']=Name
MBD['Ratings']=Ratings
MBD['Votes']=Votes
MBD['Genre']=Genre
MBD['Run_time']=Run_time
MBD['Year']=Year
MBD
```

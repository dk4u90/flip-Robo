```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser and Opening the homepage of images.google.com
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.cars24.com/buy-used-car/?sort=P&storeCityId=132&pinId=201001')
```

```python
# 500 time we scroll down by 10000 in order to generate more images on the website
for _ in range(500):
    driver.execute_script("window.scrollBy(0,10000)")
```

```python
#Make empty lists  
car_Brand=[]
car_owner=[]
car_kilometers=[]
car_fuel_type=[]
car_registration=[]
car_price=[]
car_model=[]
car_transmission=[]
# Extracting Brand from xpath
try:
    for i in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
        car_Brand.append(i.text)
except NoSuchElementException:
    car_Brand.append("_")
        
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[2]"):
        car_owner.append(i.text)
except NoSuchElementException:
    car_owner.append('-')
    
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[1]"):
        car_kilometers.append(i.text)
except NoSuchElementException:
    car_kilometers.append('-')
    
    
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='cvakB']/span"):
        car_transmission.append(i.text)
except NoSuchElementException:
    car_transmission.append('-')
    
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[3]"):
        car_fuel_type.append(i.text)
except NoSuchElementException:
    car_fuel_type.append('-')
    
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='cvakB']"):
        car_model.append(i.text)
except NoSuchElementException:
    car_model.append('-')        
        
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_19zc5']/div[2]"):
        car_price.append(i.text)
except NoSuchElementException:
    car_price.append('-')
```

```python
Car=pd.DataFrame({})
Car["car_Brand"]=car_Brand
Car["car_owner"]=car_owner
Car["car_kilometers"]=car_kilometers
Car["car_fuel_type"]=car_fuel_type
Car["car_transmission"]=car_transmission
Car["car_model"]=car_model
Car["car_price"]=car_price
Car
```

```python
# saving the dataframe
Car.to_csv('FristCar24.csv')
```

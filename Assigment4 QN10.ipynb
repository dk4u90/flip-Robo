```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Amazon.in
driver.get('https://archive.ics.uci.edu/')
```

```python
driver.find_element_by_xpath("//span[@class='whitetext'][2]/a").click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
Instances=[]
Attribute=[]
Year=[]

#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td/table/tbody/tr/td[2]/p/b"):
        Dataset_name.append(i.text)
except NoSuchElementException:
    Dataset_name.append("_")
    
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]"):
        Data_type.append(i.text)
except NoSuchElementException:
    Data_type.append("_")
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]"):
        Task.append(i.text)
except NoSuchElementException:
    Task.append("_")
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]"):
        Attribute_type.append(i.text)
except NoSuchElementException:
    Attribute_type.append("_")
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]"):
        Instances.append(i.text)
except NoSuchElementException:
    Instances.append("_")
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]"):
        Attribute.append(i.text)
except NoSuchElementException:
    Attribute.append("_")
    
#scraping Citizenship
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]"):
        Year.append(i.text)
except NoSuchElementException:
    Year.append("_")
```

```python
UCI=pd.DataFrame({})
UCI['Dataset_name']=Dataset_name
UCI['Data_type']=Data_type[1:589]
UCI['Task']=Task[1:589]
UCI['Attribute_type']=Attribute_type[1:589]
UCI['Instances']=Instances[1:589]
UCI['Attribute']=Attribute[1:589]
UCI['Year']=Year[1:589]
```

1. Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Wikipedia.
driver.get('https://en.wikipedia.org/')
```

```python
driver.find_element_by_id('searchInput').send_keys('List of most-viewed YouTube videos')
```

```python
driver.find_element_by_xpath("//input[@class='searchButton']").click()
```

```python
#Make empty lists
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]

# Extracting Rank from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[1]'):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("_")


# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[2]'):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("_")
    

# Extracting Artist from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[3]'):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("_")


# Extracting Upload_date from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[4]'):
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append("_")
    
    
# Extracting Views from xpath
try:
    for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[5]'):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("_")

```

```python
# Make dataframe
YouTube=pd.DataFrame({})
YouTube['Rank']=Rank
YouTube['Name']=Name
YouTube['Artist']=Artist
YouTube['Upload_date']=Upload_date
YouTube['Views']=Views
YouTube
```

```python
# Close current Browser
driver.close()
```

Q2. Scrape the details team India’s international fixtures from bcci.tv

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of bcci.
driver.get('https://www.bcci.tv/')
```

```python
x=driver.find_element_by_xpath('//div[@class="navigation__menu-wrapper"]/nav/ul/li[1]/div[2]/div/ul/li[1]/a').get_attribute('href')
driver.get(x)
```

```python
URL=[]
for i in driver.find_elements_by_xpath('//div[@class="event-list__list"]/div/a'):
    URL.append(i.get_attribute('href'))
```

```python
#Make empty lists
Match_title=[]
Series_Place=[]
Date=[]
Time=[]
# Scraping data from each url
for i in URL:
    # Loading the webpage by url 
    driver.get(i)
    
    # Extracting Match_title from xpath
    try:
        match_title=driver.find_element_by_xpath("//span[@class='mc-header-info__match-description-match-type mc-theme-color']")
        Match_title.append(match_title.text)
    except NoSuchElementException:
        Match_title.append("_")
        
        
    # Extracting Series_Place from xpath
    try:
        series_Place=driver.find_element_by_xpath("//h1[@class='mc-header-info__title']")
        Series_Place.append(series_Place.text)
    except NoSuchElementException:
        Series_Place.append("_")
        
    # Extracting Date from xpath
    try:
        date=driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']")
        Date.append(date.text)
    except NoSuchElementException:
        Date.append("_")
        
    # Extracting Time from xpath    
    try:
        time=driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/span")
        Time.append(time.text)
    except NoSuchElementException:
        Time.append("_")
```

```python
# Make dataframe
ICIC=pd.DataFrame({})
ICIC['Match_title']=Match_title
ICIC['Series_Place']=Series_Place
ICIC['Date']=Date
ICIC['Time']=Time
ICIC
```

```python
# Close current Browser
driver.close()
```

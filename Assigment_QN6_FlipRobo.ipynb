 Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:


```python
from bs4 import BeautifulSoup
import requests
```

i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating. 

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
t1 = soup.find_all('span', class_='u-hide-phablet')
Team_Name = []
for i in t1:
  Team_Name.append(i.text)
Team_Name
```

```python
m1 = soup.find_all('td', class_='rankings-block__banner--matches')
Matches = []
for i in m1:
  Matches.append(i.text)
Matches
```

```python
m1 = soup.find_all('td', class_='table-body__cell u-center-text')
k = []
for i in range(0,len(m1),2):
  k.append(m1[i])
for i in k:
  Matches.append(i.get_text())
Matches
```

```python
p1 = soup.find_all('td', class_='rankings-block__banner--points')
Point = []
for i in p1:
  Point.append(i.text) 
Point
```

```python
p1 = soup.find_all('td', class_='rankings-block__banner--points')
k = []
for i in range(1,len(m1),2):
  k.append(m1[i])
for i in k:
  Point.append(i.get_text())
Point
```

```python
r1 = soup.find_all('td', class_='rankings-block__banner--rating u-text-right')
Rating = []
for i in r1:
  Rating.append(i.text.strip('\n'))
  Ratings = [] 
  Ratings.append(Rating[0].strip())
Ratings
```

```python
r1 = soup.find_all('td', class_='table-body__cell u-text-right rating')
for i in r1:
  Ratings.append(i.get_text().strip('\n'))
Ratings
```

```python
import pandas as pd
ODI_Women_Team = pd.DataFrame({})
ODI_Women_Team['Team_Name'] = Team_Name[0:10]
ODI_Women_Team['Matches'] = Matches[0:10]
ODI_Women_Team['Ratings'] = Ratings[0:10]
ODI_Women_Team['Point'] = Point[0:10]
ODI_Women_Team
```

ii) Top 10 women’s ODI players along with the records of their team and rating.

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
p1=soup.find_all('div',class_='rankings-block__banner--name-large')
Player = []
for i in p1:
  Player.append(i.text)
Player
```

```python
p2 = soup.find_all('td', class_='table-body__cell rankings-table__name name')
for i in p2:
  for j in i.find_all("a"):
    Player.append(j.text)
Player
```

```python
t1 = soup.find("div", class_="rankings-block__banner--nationality")
Team = []
for i in t1:
  Team.append(t1.text.strip())
```

```python
t2 = soup.find_all("span", class_='table-body__logo-text')
for i in t2:
  Team.append(i.text)
Team
```

```python
r1 = soup.find_all('td', class_='table-body__cell rating')
Rating = []
for i in r1:
  Rating.append(i.text)
Rating
```

```python
import pandas as pd
ODI_Women_Players=pd.DataFrame({})
ODI_Women_Players['Player']=Player[0:10]
ODI_Women_Players['Team']=Team[0:10]
ODI_Women_Players['Rating']=Rating[0:10]
ODI_Women_Players
```

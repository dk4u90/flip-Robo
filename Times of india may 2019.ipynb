```python
# Importing important Libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
# Opening the homepage of Wikipedia.
driver.get('https://timesofindia.indiatimes.com/archive/year-2019,month-5.cms')
```

```python
Date_URL=[]
try:
    #scraping Date_URL
    for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr/td/a"):
        #appending the urls in Date_URL list
        Date_URL.append(i.get_attribute('href'))
except NoSuchElementException :
    Date_URL.append("Not Available")
```

```python
Date_URL
```

### Upper three URL is same for 1 May 2019 beacuse of scrapped data from [3:34] 

```python
Headline_URL=[]
for j in Date_URL[0:1]:
    driver.get(j)
    try:
        #scraping Headline_URL
        for i in driver.find_elements_by_xpath("//div[@style='font-family:arial ;font-size:12;font-weight:bold; color: #006699']/table/tbody/tr[2]/td/span/a"):
            #appending the urls in Headline_URL list
            Headline_URL.append(i.get_attribute('href'))
    except NoSuchElementException :
        Headline_URL.append("Not Available")
```

```python
Headline_URL
```

### Only 10 Headline has been scrapped due to taking long time 

```python
#for loop for scrapping data.
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
for url in Headline_URL[0:10]:
    driver.get(url)
    time.sleep(2)
    # Extracting Headline from xpath
    try: 
        headline=driver.find_element_by_xpath("//div[@class='as_heading']/h1").text
        Headline.append(headline)
    except NoSuchElementException :
        Headline.append("Not Available")
        
    # Extracting Vertical from xpath       
    try: 
        vertical=driver.find_element_by_xpath("//a[@class='subsecactive ']").text
        Vertical.append(vertical)
    except NoSuchElementException :
        Vertical.append("Not Available")
        
    # Extracting Author from xpath
    try: 
        author=driver.find_element_by_xpath("//div[@class='as_heading']/div[2]/div[2]").text
        Author.append(author)
    except NoSuchElementException :
            Author.append("Not Available")
            
    # Extracting Date from xpath
    try: 
        date=driver.find_element_by_xpath("//div[@class='as_heading']/div[2]/div[3]").text
        Date.append(date)
    except NoSuchElementException :
        Date.append("Not Available")
            
    # Extracting Description from xpath
    try: 
        description=driver.find_element_by_xpath("//div[@class='Normal']").text
        Description.append(description)
    except NoSuchElementException :
        Description.append("Not Available")
```

```python
#create Dataframe.
Times_of_India=pd.DataFrame({})
Times_of_India['Date']=Date
Times_of_India['Author']=Author
Times_of_India['Vertical']=Vertical
Times_of_India['Headline']=Headline
Times_of_India['Description']=Description
Times_of_India
```

```python
# saving the dataframe
Times_of_India.to_csv('TOI.csv')
```

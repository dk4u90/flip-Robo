```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.amazon.in/s?k=mobile&rh=p_36%3A-20000000&qid=1621975129&rnid=1318502031&ref=sr_nr_p_36_5")
page.status_code
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
soup
```

```python
m_name = soup.find_all('span', class_='a-size-medium a-color-base a-text-normal')
m_price = soup.find_all('span', class_='a-price-whole')
m_rating = soup.find_all('span', class_="a-icon-alt")
m_img = soup.find_all('img', class_="s-image")
```

```python
Mobile = []
for i in m_name:
 Mobile.append(i.get_text())
Mobile        
```

```python
Price = []
for i in m_price:
 Price.append(i.get_text())
Price
```

```python
        
Rating = []
for i in m_rating:
 Rating.append(i.get_text())
Rating
```

```python
        
Image = []
for i in m_img:
 Image.append(i.get('srcset'))
Image
```

```python
import pandas as pd

Mobile_dataset = pd.DataFrame()
Mobile_dataset['Mobile_Name'] = Mobile[0:15]
Mobile_dataset['Mobile_Price'] = Price[0:15]
Mobile_dataset['Mobile_rating'] = Rating[0:15]
Mobile_dataset['Image_Link'] = Image[0:15]
Mobile_dataset
```
